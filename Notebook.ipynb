### Cleaning Data

```python
# Mendefinisikan data yang sulit dimengerti
season_codes = {1: 'spring', 2:'summer', 3:'fall', 4: 'winter'}
year_codes = {0: 2011, 1: 2012}
month_codes = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'June',7:'July',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
weekday_codes = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
working_day_codes = {1:'working day',0:'holiday'}
weather_situation_codes = {1:'Clear',2:'Mist',3:'Light Snow',4:'Heavy Rain'}
```

#### Membersihkan data pada tabel `day`

```python
# Menghapus column 'instant' dan 'holiday' karena tidak digunakan
day_df.drop(['instant', 'holiday'], axis=1, inplace=True)
```

```python
#  Merubah nama column untuk lebih mudah memahami data
day_df.rename(columns={'dteday':'date','yr':'year','mnth':'month','weekday':'week_day',
                   'workingday':'working_day','weathersit':'weather_situation','atemp':'temp_feel',
                   'hum':'humidity','windspeed':'wind_speed','cnt':'count'},inplace=True)
day_df.head()
```

```python
# Merubah data season agar lebih mudah dimengerti
day_df['season'] = day_df['season'].map(season_codes)
day_df.season.value_counts()
```

```python
# Merubah data year agar lebih mudah dimengerti
day_df['year'] = day_df['year'].map(year_codes)
day_df.year.value_counts()
```

```python
# Merubah data month agar lebih mudah dimengerti
day_df['month'] = day_df['month'].map(month_codes)
day_df.month.value_counts()
```

```python
# Merubah data weekday agar lebih mudah dimengerti
day_df['week_day'] = day_df['week_day'].map(weekday_codes)
day_df.week_day.value_counts()
```

```python
# Merubah data working_day agar lebih mudah dimengerti
day_df['working_day'] = day_df['working_day'].map(working_day_codes)
day_df.working_day.value_counts()
```

```python
# Merubah data weeather_situation agar lebih mudah dimengerti
day_df['weather_situation'] = day_df['weather_situation'].map(weather_situation_codes)
day_df.weather_situation.value_counts()
```

```python
day_df.head()
```

#### Membersihkan data pada tabel `hour`

```python
# Menghapus column 'instant' karena tidak digunakan
hour_df.drop(['instant', 'holiday'], axis=1, inplace=True)
```

```python
#  Merubah nama column untuk lebih mudah memahami data
hour_df.rename(columns={'dteday':'date','yr':'year','mnth':'month', 'hr': 'hour' ,'weekday':'week_day',
                   'workingday':'working_day','weathersit':'weather_situation','atemp':'temp_feel',
                   'hum':'humidity','windspeed':'wind_speed','cnt':'count'},inplace=True)
hour_df.head()
```

```python
# Merubah data season agar lebih mudah dimengerti
hour_df['season'] = hour_df['season'].map(season_codes)
hour_df.season.value_counts()
```

```python
# Merubah data year agar lebih mudah dimengerti
hour_df['year'] = hour_df['year'].map(year_codes)
hour_df.year.value_counts()
```

```python
# Merubah data month agar lebih mudah dimengerti
hour_df['month'] = hour_df['month'].map(month_codes)
hour_df.month.value_counts()
```

```python
# Merubah data week day agar lebih mudah dimengerti
hour_df['week_day'] = hour_df['week_day'].map(weekday_codes)
hour_df.week_day.value_counts()
```

```python
# Merubah data working day agar lebih mudah dimengerti
hour_df['working_day'] = hour_df['working_day'].map(working_day_codes)
hour_df.working_day.value_counts()
```

```python
# Merubah data weather situation day agar lebih mudah dimengerti
hour_df['weather_situation'] = hour_df['weather_situation'].map(weather_situation_codes)
hour_df.weather_situation.value_counts()
```
